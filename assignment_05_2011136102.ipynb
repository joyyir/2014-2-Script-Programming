###2. 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

```python
class Person:
    def sayHello(self):
        print '...'
        
class Korean(Person):
    def sayHello(self):
        print '안녕'
        
class American(Person):
    def sayHello(self):
        print 'Hello'
        
class Chinese(Person):
    def sayHello(self):
        print '你好.'

for each in (Korean(), American(), Chinese()):
    each.sayHello()
```

####설명
- 다형성이란, 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능이다.  
- 위의 예제에서는 Person으로부터 상속을 받은 Korean, American, Chinese 클래스의 인스턴스들이 sayHello()라는 멤버 함수 호출에 대해 각각 다르게 반응하고 있음을 볼 수 있다.

###3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오. (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)
  - 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
    - count: Counter 인스턴스가 지니는 초기 정수 값
    - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)

    > \>\>\> c = Counter(10) <br/>
    > \>\>\> d = Counter(10, 2) <br/>
    
  - 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 \_\_str\_\_() 메소드를 구현하시오.
    > \>\>\> print c  <br/>
    > [Count (step: 1)] 10 <br/> 
    > \>\>\> print d  <br/>
    > [Count (step: 2)] 10 <br/>
    
  - 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 구현하시오.
    > \>\>\> c.incr()  <br/>

  - 요구사항 4. \_\_call\_\_() 메소드를 추가하여 인스턴스 객체를 직접 호출할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때의 동작방식은 incr() 메소드를 호출하였을 때와 동일하다. 
    > \>\>\> c()  <br/>
    > \>\>\> d()  <br/>
    > \>\>\> print c  <br/>
    > [Count (step: 1)] 12 <br/> 
    > \>\>\> print d  <br/>
    > [Count (step: 2)] 14 <br/>
    
  - 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 관련 메소드를 추가하시오.
    > \>\>\> c = c + 5  <br/>
    > \>\>\> d = d - 5  <br/>
    > \>\>\> print c  <br/>
    > [Count (step: 1)] 17 <br/> 
    > \>\>\> print d  <br/>
    > [Count (step: 2)] 9 <br/>  
    
  - 요구사항 6. 다음과 같은 관계연산 (+, -)이 수행될 수 있도록 \_\_cmp\_\_() 메소드를 추가하시오.
    > \>\>\> print c > 10  <br/>
    > True <br/> 
    > \>\>\> print d > 10 <br/>
    > False <br/>
    > \>\>\> print c < 10  <br/>
    > False <br/> 
    > \>\>\> print d < 10 <br/>
    > True <br/>
    > \>\>\> print c == 10  <br/>
    > True <br/> 
    > \>\>\> print d != 9 <br/>
    > False <br/>

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
        
    def __str__(self):
        return '[Count (step: %d)] %d' % (self.step, self.count)
        
    def incr(self):
        self.count += self.step
        
    def __call__(self):
        self.incr()
        
    def __add__(self, x):
        return Counter(self.count + x, self.step)
        
    def __sub__(self, x):
        return Counter(self.count - x, self.step)
        
    def __cmp__(self, other):
        return self.count - other;

    
    
c = Counter(10)
d = Counter(10, 2)
print
print '* 요구사항 2'
print c
print d
print
print '* 요구사항 3'
c.incr()
d.incr()
print c
print d
print
print '요구사항 4'
c()
d()
print c
print d
print
print '* 요구사항 5'
c = c + 5
d = d - 5
print c
print d
print
print '* 요구사항 6'
print c > 10
print d > 10
print c < 10
print d < 10
print c == 10
print d != 9
```

###4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init\_\_(), \_\_str()\_\_(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
  - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

####설명
    먼저 생성자 메서드부터 보겠습니다. 현재 MySet 클래스는 list 클래스로부터 상속을 받았습니다. 그러면 list 클래스의 모든 변수와 함수를 물려받았을 것입니다. 그러면 self는 append 함수를 쓸 수 있게 됩니다. 먼저 l에 있는 값들을 self에 넣은 다음, 정적 함수인 eliminate_duplicate 함수를 호출합니다. 이 함수는 정적 함수여서 인스턴스 없이도 클래스 이름으로 호출이 가능합니다. 이 함수를 호출함으로써 self의 중복된 값들이 제거됩니다. 눈여겨 볼 점은, l[:] = [] 소스를 이용해서 l의 id를 유지했다는 점입니다. l = [] 같은 소스를 썼다면, 제대로 동작하지 않을 것입니다. 마지막으로 __str__ 함수를 보겠습니다. 이 함수는 함수 내용을 string으로 변환시켜주는 함수입니다. 데이터는 현재 리스트에 저장이 되어있지만 문자열로 바꿀 때는 {} 안에 데이터가 들어가는 형태로 만들었습니다.

###5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오 (정답을 각 요구사항별로 입력할 필요 없이 5번 문제에 대해 1개의 MySet 클래스 정의 코드를 제시하면 된다.)
  - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
    > \>\>\> s = s | t  <br/>
    > \>\>\> print s  <br/>
    > MySet: {1 ,2 ,3 ,4 ,5 ,6 ,7 ,8 ,9} <br/> 
    
  - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> s = s & t  <br/>
    > \>\>\> print s  <br/>
    > MySet: {3} <br/>

  - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> s = s - t  <br/>
    > \>\>\> print s  <br/>
    > MySet: {1, 2} <br/>

```python
# -*- coding: utf-8 -*-
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, other):
        result = MySet([])
        for ele in self:
            result.append(ele)
        for ele in other:
            if ele not in result:
                result.append(ele)
        return result
        
    def __and__(self, other):
        result = MySet([])
        for ele in self:
            if ele in other:
                result.append(ele)
        return result

    def __sub__(self, other):
        result = MySet([])
        for ele in self:
            if ele not in other:
                result.append(ele)
        return result        

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    print
    print '* 합집합'
    s = s | t
    print s
    print
    print '* 교집합'
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    s = s & t
    print s
    print
    print '* 차집합'
    s = MySet([1, 2, 3]) 
    t = MySet([3, 4, 5]) 
    s = s - t 
    print s
```

####설명
- 합집합 : 빈 MySet 클래스 인스턴스인 result를 만들고, self 값들을 다 집어넣는다. 그리고나서 other 값들을 하나하나 result에 이미 있는지 비교해보고, 없다면 result에 집어넣는다. result를 반환한다.
- 교집합 : 빈 MySet 클래스 인스턴스인 result를 만들고, self의 값들을 하나하나 other에도 있는지 비교해보고, other에도 있다면 result에 집어넣는다. result를 반환한다.
- 차집합 : 빈 MySet 클래스 인스턴스인 result를 만들고, self의 값들을 하나하나 other에도 있는지 비교해보고, other에 없다면 result에 집어넣는다. result를 반환한다.

```python
# -*- coding: utf-8 -*-

############## 기존 소스 시작 ###############

import urllib
import string

def getWordFrequencyDic(url):
    s = urllib.urlopen(url).read()
    temp_result = []
    open_cnt = 0
    close_cnt = 0
    is_in_par = False;
    is_style_script = False;
    range_s = range(len(s));
    for k in range_s:
   	if is_in_par:
  		if s[k] == '>':
 			close_cnt = close_cnt + 1
 			if s.startswith('/style', k-6) or s.startswith('/STYLE', k-6) or\
 			s.startswith('/script', k-7) or s.startswith('/SCRIPT', k-7) :
    				is_style_script = False;
 			if open_cnt == close_cnt:
    				is_in_par = False
 			continue
  		elif s[k] == '<':
 			open_cnt = open_cnt + 1
  		else:
 			continue
   	else: # not in par
  		if s[k] == '<':
 			temp_result.append(' ')
 			open_cnt = open_cnt + 1
 			is_in_par = True
 			if s.startswith('style', k+1) or s.startswith('STYLE', k+1) or\
 			s.startswith('script', k+1) or s.startswith('SCRIPT', k+1) :
    				is_style_script = True;
 			continue
  		else:
 			if is_style_script:
    				continue
 			else:
    				temp_result.append(s[k])
    result = ''.join(temp_result)
    
    # 단어의 구두점 제거 및 길이가 0인 단어 제거
    dic = result.split()
    new_dic = []
    for word in dic:
        word_temp = []
        word_temp2 = ''
        for ch in word:
            if ch not in string.punctuation:
                word_temp.append(ch)
        word_temp2 = ''.join(word_temp)
        #word_temp2.strip('\t\r\n ')
        if len(word_temp2) == 0:
            continue
        else:
            new_dic.append(word_temp2)
    
    # 사전으로 재구성
    new_result = {}
    for word in new_dic:
        new_result[word] = new_dic.count(word)
    
    return new_result
    # 함수 정의 끝
    
############## 기존 소스 끝 ###############

class WebWordsFrequency:
    def __init__(self, *arg):
        self.urlList = []
        for url in arg:
            self.urlList.append(url)
            
    def addUrl(self, url):
        self.urlList.append(url)
        
    def removeUrl(self, url):
        self.urlList.remove(url)
        
    def listUrls(self):
        for url in self.urlList:
            print url
    
    def getWordsFrequency(self):
        resultDic = {} # 결과를 저장할 빈 사전 생성
        for url in self.urlList:
            wordsDic = getWordFrequencyDic(url) # url로부터 단어 사전을 추출
            for word in wordsDic:
                if word in resultDic: # 기존에 단어가 존재하면,
                    resultDic[word] += wordsDic[word] # 개수 추가
                else: # 기존에 단어가 존재하지 않다면,
                    resultDic[word] = wordsDic[word] # 개수 입력
        return resultDic
    
    def getMaxFreqencyWords(self):
        resultList = [] # 결과를 저장할 빈 리스트 생성
        sortedDic = sorted(self.getWordsFrequency().items(), key=lambda x:x[1], reverse=True) # 사전을 튜플로 구성된 리스트로 변환 후, 내림차순 정렬 수행
        if len(sortedDic) > 0:
            maxFrequency = sortedDic[0][1] # 최대 빈도수
        for tup in sortedDic:
            if tup[1] != maxFrequency: # 최대 빈도수보다 작으면 끝내라.
                break
            else: # 최대 빈도수인 단어를 resultList에 추가한다.
                resultList.append(tup[0])
                #print '%s - %d번' % tup
        return resultList
                
# __main__

w1 = WebWordsFrequency('http://www.daum.net', 'http://www.naver.com', 'http://www.google.co.kr')
w2 = WebWordsFrequency('http://www.daum.net', 'http://www.naver.com')    
w3 = WebWordsFrequency() 

w1.addUrl('http://cse.koreatech.ac.kr')
w3.addUrl('http://www.koreatech.ac.kr')

w1.removeUrl('http://www.daum.net')
w2.removeUrl('http://www.naver.com')

print '* w1.listUrls()'
w1.listUrls()
print
print
print '* w1.getWordsFrequency()'
print '{',;
for tup in w1.getWordsFrequency().items():
    print "'%s':%d," % tup,;
print '}'
print
print
print '* w1.getMaxFreqencyWords()'
for word in w1.getMaxFreqencyWords(): 
    print word
print
print
print '* w2.getMaxFreqencyWords()'
for word in w2.getMaxFreqencyWords(): 
    print word
```

####설명
- 생성자 메서드에서는 URL들을 저장하기위한 urlList 리스트를 초기화한다.
- addUrl에서는 인자로 받은 url을 urlList에 추가한다.
- removeUrl에서는 urlList에서 인자로 받은 url을 제거한다.
- listUrl에서는 urlList의 내용을 출력한다.
- getWordsFrequency에서는 Assignment 3에서 단어와 빈도수가 저장된 사전을 만드는 함수를 이용한다. resultDic이라는 빈 사전 안에 각 URL으로부터 추출된 사전을 추가한다. 이 때, resultDic에 이미 단어가 있으면 사전의 value에 단어의 수만큼을 더한다.
- getMaxFrequencyWords에서는 위에서 만든 getWordsFrequency 함수로부터 반환되는 사전을 리스트로 바꾸어 정렬한 다음에, 최대 빈도수를 갖는 단어를 출력하게 한다.

###8. 7번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여 OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.
- 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.  
    
> \>\>\> w4 = OrderedWebWordsFrequency('http://www.daum.net', 'http://www.naver.com', 'http://www.google.co.kr') <br/>
> \>\>\> w4.getWordsFrequency() <br/>
> [('안녕', 12), ('우리', 11), ('다음', 8), ('world', 2), ('hello', 1), ('python', 1)]

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self):
        resultDic = {}
        for url in self.urlList:
            wordsDic = getWordFrequencyDic(url)
            for word in wordsDic:
                if word in resultDic:
                    resultDic[word] += wordsDic[word]
                else:
                    resultDic[word] = wordsDic[word]
        sortedDic = sorted(resultDic.items(), key=lambda x:x[1], reverse=True)
        print '[',
        for tup in sortedDic:
            print '(\'%s\', %d),' % tup,
        print ']'
        
        
w4 = OrderedWebWordsFrequency('http://www.daum.net', 'http://www.naver.com', 'http://www.google.co.kr') 
print
print
print '* w4.getMaxFreqencyWords()'
w4.getWordsFrequency()
```
